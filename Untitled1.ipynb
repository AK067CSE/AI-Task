```python
!pip install -q transformers pillow requests replicate

import torch
from diffusers import StableDiffusionPipeline
from PIL import Image, ImageDraw, ImageFont
from transformers import pipeline

def extract_key_points(text):
    s = pipeline("summarization", model="facebook/bart-large-cnn")
    return s(text, max_length=130, min_length=30, do_sample=False)[0]['summary_text'].split(". ")[:5]

def generate_prompts(points):
    return [f"Healthcare infographic, minimal style, blue color theme, icons, about: {p}" for p in points]

def generate_image(prompt, idx):
    pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16, use_auth_token=True)
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
    img = pipe(prompt, guidance_scale=7.5).images[0]
    path = f"visual_{idx}.png"; img.save(path)
    return path

def overlay_text(img_path, text, out_path):
    img = Image.open(img_path).convert("RGBA")
    draw = ImageDraw.Draw(img); font = ImageFont.load_default()
    pos = (10, img.height - 60); area = (pos[0], pos[1], img.width - 10, img.height - 10)
    draw.rectangle(area, fill=(255, 255, 255, 220)); draw.text(pos, text, font=font, fill="black")
    img.save(out_path)

text = """Digital healthcare in 2025 will undergo a massive transformation. AI diagnostics are improving accuracy by 40%. Wearables will become standard for real-time monitoring. Telemedicine adoption is expected to cover 75% of remote consultations. Blockchain will secure patient data. Personalized medicine will be driven by genomics."""
points = extract_key_points(text)
prompts = generate_prompts(points)

out_files = []
for i, (p, pr) in enumerate(zip(points, prompts)):
    raw = generate_image(pr, i)
    final = f"final_visual_{i}.png"; overlay_text(raw, p[:90], final); out_files.append(final)

print("All visuals generated successfully.")
for f in out_files: display(Image.open(f))

with open("key_points.txt", "w") as f: f.write("\n".join(points))
```
